```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json("rag_ollama_evaluation.json")

collapsed_metrics = [j["metrics_data"] for i in df["test_results"].tolist() for j in i]

len(collapsed_metrics[0])
```

```python
context_relevancy = []
context_precision = []
context_recall = []
answer_relevancy = []
faithfulness = []
for item in collapsed_metrics:
    context_relevancy.append(
        [m["score"] for m in item if m["name"] == "Contextual Relevancy"][0]
    )
    context_precision.append(
        [m["score"] for m in item if m["name"] == "Contextual Precision"][0]
    )
    context_recall.append(
        [m["score"] for m in item if m["name"] == "Contextual Recall"][0]
    )
    answer_relevancy.append(
        [m["score"] for m in item if m["name"] == "Answer Relevancy"][0]
    )
    faithfulness.append(
        [m["score"] for m in item if m["name"] == "Faithfulness"][0]
    )


print("Contextual Relevancy: ", np.mean(context_relevancy))
print("Contextual Precision: ", np.mean(context_precision))
print("Contextual Recall: ", np.mean(context_recall))
print("Answer Relevancy: ", np.mean(answer_relevancy))
print("Faithfulness: ", np.mean(faithfulness))

```
